    - Build a program to manage student grades using lists and tuples

```python
# List to store student records
students = []

# Ask number of students
n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter student name: ")
    grade = float(input("Enter student grade: "))

    # Store as tuple (name, grade)
    student_record = (name, grade)

    # Add tuple to list
    students.append(student_record)

print("\n Student Records:")
for student in students:
    print(f"Name: {student[0]}, Grade: {student[1]}")

# Calculate class average
total = sum(student[1] for student in students)
average = total / len(students)

print(f"\nClass Average Grade = {average:.2f}")
```
