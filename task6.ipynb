```python
# Get user input
first_name = input("Enter your first name: ").strip()
last_name = input("Enter your last name: ").strip()
fav_word = input("Enter your favourite word: ").strip()
part1 = first_name[:3].lower()
part2 = last_name[-2:].upper()
part3 = fav_word.replace(" ", "_").title()
part4 = str(len(first_name) + len(last_name))

username = f"{part1}{part3}{part2}{part4}"

print("\nâœ… Your generated username is:", username)
```
