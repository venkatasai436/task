- 1. Create a personal library management system that tracks your book collection using only dictionaries. 
- 2.  This task will help you practice all major dictionary operations and methods.

- Task Description

    - Build a program that manages a personal book library with the following features:
        - Add new books to your library
        - Remove books from your library
        - Search for books by various criteria
        - Update book information
        - Generate reports about your library

Calculate statistics about your collection
Requirements

- You must use ONLY dictionaries (no lists, sets, or other data structures for the main operations). You can use basic data types (strings, integers, booleans) and control structure

```python

ibrary = {}   # Stores all books (book_id: book_info)
next_id = 1    # To generate unique book IDs

while True:
    print("\n1. Add Book")
    print("2. View Books")
    print("3. Search Book by Title")
    print("4. Delete Book")
    print("5. Exit")

    choice = input("Enter choice: ")

    #  Add Book
    if choice == "1":
        global next_id
        title = input("Enter title: ")
        author = input("Enter author: ")
        year = input("Enter year: ")

        library[str(next_id)] = {
            "title": title,
            "author": author,
            "year": year
        }
        print(f"Book added with ID {next_id}")
        next_id += 1

    #  View All Books
    elif choice == "2":
        if not library:
            print("No books yet!")
        else:
            for book_id, info in library.items():
                print(f"ID: {book_id} → {info}")

    #  Search Book
    elif choice == "3":
        search_title = input("Enter title to search: ").lower()
        found = False
        for book_id, info in library.items():
            if search_title in info["title"].lower():
                print(f"Found → ID: {book_id}, {info}")
                found = True
        if not found:
            print("No matching book found.")

    #  Delete Book
    elif choice == "4":
        remove_id = input("Enter book ID to delete: ")
        if remove_id in library:
            library.pop(remove_id)
            print("Book removed successfully!")
        else:
            print("Invalid ID!")

    #  Exit
    elif choice == "5":
        print("Goodbye!")
        break

    else:
        print("Invalid choice!")
```
